# FPGA Floating state

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load all three files
pbit1_path = "FPGA Data/pbit1flt.Wfm.csv"
pbit2_path = "FPGA Data/pbit2flt.Wfm.csv"
pbit3_path = "FPGA Data/pbit3flt.Wfm.csv"

pbit1_df = pd.read_csv(pbit1_path)
pbit2_df = pd.read_csv(pbit2_path)
pbit3_df = pd.read_csv(pbit3_path)

# Combine the columns into a single DataFrame
combined_df = pd.concat([pbit1_df, pbit2_df, pbit3_df], axis=1)
combined_df.columns = ["pbit1flt", "pbit2flt", "pbit3flt"]

# Convert to binary state strings
combined_df["binary_state"] = combined_df.astype(str).agg("".join, axis=1)

# Count occurrences of each unique binary state
state_counts = combined_df["binary_state"].value_counts()/len(combined_df)
print(f"Sample Size = {len(combined_df)}")

# Define the expected binary state order
expected_order = ["000", "001", "010", "011", "100", "101", "110", "111"]

# Ensure all states are present (fill missing with 0 count)
state_counts = state_counts.reindex(expected_order, fill_value=0)

# Plot with the correct order
plt.figure(figsize=(8, 5))
plt.bar(state_counts.index, state_counts.values)
plt.xlabel("Binary State")
plt.ylabel("Count")
plt.title("Occurrences of Unique Binary States (Ordered)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Combine all files into one with corresponding label

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define file paths
pbit1flt = "FPGA Data/pbit1flt.Wfm.csv"
pbit2flt = "FPGA Data/pbit2flt.Wfm.csv"
pbit3flt = "FPGA Data/pbit3flt.Wfm.csv"

pbit1a1b1 = "FPGA Data/pbit1a1b1.Wfm.csv"
pbit2a1b1 = "FPGA Data/pbit2a1b1.Wfm.csv"
pbit3a1b1 = "FPGA Data/pbit3a1b1.Wfm.csv"

pbit1c0 = "FPGA Data/pbit1c0.Wfm.csv"
pbit2c0 = "FPGA Data/pbit2c0.Wfm.csv"
pbit3c0 = "FPGA Data/pbit3c0.Wfm.csv"

# Reload the files
pbit1flt_df = pd.read_csv(pbit1flt)
pbit2flt_df = pd.read_csv(pbit2flt)
pbit3flt_df = pd.read_csv(pbit3flt)

pbit1a1b1_df = pd.read_csv(pbit1a1b1)
pbit2a1b1_df = pd.read_csv(pbit2a1b1)
pbit3a1b1_df = pd.read_csv(pbit3a1b1)

pbit1c0_df = pd.read_csv(pbit1c0)
pbit2c0_df = pd.read_csv(pbit2c0)
pbit3c0_df = pd.read_csv(pbit3c0)

# Combine the columns into separate DataFrames
combined_flt_df = pd.concat([pbit1flt_df, pbit2flt_df, pbit3flt_df], axis=1)
combined_a1b1_df = pd.concat([pbit1a1b1_df, pbit2a1b1_df, pbit3a1b1_df], axis=1)
combined_c0_df = pd.concat([pbit1c0_df, pbit2c0_df, pbit3c0_df], axis=1)

combined_flt_df.columns = ["pbit1", "pbit2", "pbit3"]
combined_a1b1_df.columns = ["pbit1", "pbit2", "pbit3"]
combined_c0_df.columns = ["pbit1", "pbit2", "pbit3"]

# Define labels
labels = ["Floating State (I0=1)", "A=1, B=1 (I0=1)", "C=0 (I0=1)"]

# Function to process data
def process_data(df, label):
    df["case"] = df.astype(str).agg("".join, axis=1)
    state_counts = df["case"].value_counts(normalize=True)
    expected_order = ["000", "001", "010", "011", "100", "101", "110", "111"]
    state_counts = state_counts.reindex(expected_order, fill_value=0)
    return pd.DataFrame({"label": label, "case": state_counts.index, "normalized_count_FPGA": state_counts.values})

# Process all datasets
final_df = pd.concat([
    process_data(combined_flt_df, labels[0]),
    process_data(combined_a1b1_df, labels[1]),
    process_data(combined_c0_df, labels[2])
], ignore_index=True)

# Save the final DataFrame (optional)
output_path = "FPGA Data/AND_gate_FPGA_imp_combined.csv"
final_df.to_csv(output_path, index=False)
```
