```python
import numpy as np
import Scripts.CustomScripts as CS
def int_to_bits(n, width):
    """Convert integer n into a list of bits of given width (MSB first)."""
    return [(n >> i) & 1 for i in reversed(range(width))]

truth_table = []
# Iterate over all 4-bit numbers for A and B.
for A in range(16):
    for B in range(16):
        for Cin in range(2):
            A_bits = int_to_bits(A, 4)   # 4 bits for A
            B_bits = int_to_bits(B, 4)   # 4 bits for B
            total = A + B + Cin          # Sum including carry-in (max = 15+15+1 = 31)
            total_bits = int_to_bits(total, 5)  # Represent result in 5 bits
            # In the 5-bit result, the most significant bit is the carry out,
            # and the remaining 4 bits are the sum bits.
            truth_table.append(A_bits + B_bits + [Cin] + total_bits)


truth_table = np.array(truth_table)
print(len(truth_table))
columns = ["A3", "A2", "A1", "A0", "B3", "B2", "B1", "B0", "Cin", "Cout", "S3", "S2", "S1", "S0"]


output_columns = ["Cout", "S3", "S2", "S1", "S0"]
CS.truth_table_probabilities(truth_table,columns,output_columns,figWidth=8)
```
