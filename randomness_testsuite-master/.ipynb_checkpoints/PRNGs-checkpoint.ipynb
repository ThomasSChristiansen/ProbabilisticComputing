```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

32 bit LFSR

```python
# Parameters
seed = 0xACB1
n = 10000000  #antal bits generreret
weight = 0x80000000    #vægt, hold den på 0x80000000 for at sandsynligheden er 50/50

taps = [8, 6, 5, 4]   #LFSR tap lokationer

tap_mask = sum([1 << (tap - 1) for tap in taps])  #eh

state = seed  #sætter start staten til at være seeded

output_bits_lfsr = []  #til at gemme output

output_preweight_lfsr = []  #til at gemme 32 bit output

for _ in range(n):
    generated_value = state  #state updateres
    
    output_preweight_lfsr.append(generated_value)  #32 bit output gemmes

    if generated_value > weight:  #state sammelignes med vægt
        output_bits_lfsr.append(1)
    else:
        output_bits_lfsr.append(0)

    feedback = 0  #xor halløj
    for tap in taps:
        feedback ^= (state >> (tap - 1)) & 1

    state = (state >> 1) | (feedback << 31)  #shift

#print("Comparison results (as bits):", ''.join(map(str, output_bits)))  #output

print('done')

#print(output_preweight_lfsr)
```

xorshiro128++

```python
# Parametre
seed = [0xACB1, 0x1234, 0x5678, 0x9ABC]  #4 seeds da xoshiro bruger 4 registre
n = 10000000                             #antal bits generreret
weight = 0x80000000                      #vægt, hold den på 0x80000000 for at sandsynligheden er 50/50

state = seed[:]  #sætter start staten til at være seeded

output_preweight_xoshiro = []  #til at gemme 32 bit output

output_bits_xoshiro = []  #til at gemme output i nuller og ettaller

def rotl(x, k):
    """Rotate left function for 32-bit integers."""
    return ((x << k) | (x >> (32 - k))) & 0xFFFFFFFF

def xoshiro128pp(state):
    """xoshiro128++ pseudorandom number generator."""
    s0, s1, s2, s3 = state
    result = (rotl(s0 + s3, 7) + s0) & 0xFFFFFFFF

    t = (s1 << 9) & 0xFFFFFFFF
    state[2] ^= s0
    state[3] ^= s1
    state[1] ^= s2
    state[0] ^= s3
    state[2] ^= t
    state[3] = rotl(state[3], 11)

    return result

for _ in range(n):
    generated_value = xoshiro128pp(state)  #generrere ny xoshiro output
    
    output_preweight_xoshiro.append(generated_value)

    if generated_value > weight:  #sammenligner med vægt
        output_bits_xoshiro.append(1)
    else:
        output_bits_xoshiro.append(0)

# Output the results
#print("Comparison results (as bits):", ''.join(map(str, output_bits)))

print('done')
```

write binary file

```python
with open(f"binary file lfsr32 with {n} bits", 'w') as e_file:
    # Write the binary string to the file
    e_file.write(''.join(str(x) for x in output_bits_lfsr))
    
with open(f"binary file xoshiro128++ with {n} bits", 'w') as e_file:
    # Write the binary string to the file
    e_file.write(''.join(str(x) for x in output_bits_xoshiro))
```
