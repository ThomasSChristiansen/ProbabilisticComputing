# Full Adder python simulation

```python
import Scripts.simulate_p_circuit as simulate_p_circuit
import Scripts.CustomScripts as CS

J, h, node_order = CS.load_npz_data("FA.npz", target_folder="Circuit_Library")

# Define your target_names dictionary
target_names = {
    "Floating State": {
    },
    "C_in=0": {
        "C_in": "push0",
    },
    "1+1": {
        "C_in": "push1",
        "A": "push1",
        "B": "push1",
    },
}

# Get the updated configurations:
configured_h = CS.update_all_configurations(target_names, h, node_order)

selected_bits = {
    "Floating State": ['C_in', 'B', 'A', 'S', 'C_out'],  
    "C_in=0": ['C_in', 'B', 'A', 'S', 'C_out'],  
    "1+1": ['C_in', 'B', 'A', 'S', 'C_out'], 
}

simulate_p_circuit.plot_probabilities(num_steps=100000, savefig=False, save_csv=False,J_bipolar=J, node_order=node_order,
                                      h_bipolar_dict=configured_h, use_python=True, use_boltzmann=False, 
                                      filename="Full_Adder_Python_Sim_5Bit", figWidth=8, selected_bits=selected_bits)
```
