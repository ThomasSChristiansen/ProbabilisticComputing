```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("CSV files\AND_UniqueOccurrences.csv", low_memory=False, dtype={'Pattern': str})
# print(df)
# print(df.head())
# print(df.info())
# Create a bar plot using the "Pattern" as the x-axis and "Occurrence" as the y-axis.
plt.figure(figsize=(8, 6))
plt.bar(df["Pattern"], df["Occurrence"]/df["Occurrence"].sum(), color='skyblue')
plt.xlabel("bit Pattern")
plt.ylabel("Occurrence")
plt.title("Output Occurrences (unoptimized)")
plt.margins(x=0.01)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
